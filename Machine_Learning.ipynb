Q.1 - Explain the differences between AI, ML, Deep Learning (DL), and Data Science (DS).
  - Artificial Intelligence (AI), Machine Learning (ML), Deep Learning (DL), and Data Science (DS) are interconnected but distinct fields. AI is the overarching concept, with ML and DL as subsets, while DS is a broader, multidisciplinary field that uses many of these techniques as tools

  The relationship visualized

A simple way to understand the hierarchy is a set of nesting dolls, with AI being the largest doll.

AI (Artificial Intelligence): The biggest doll, representing the overall goal of creating intelligent machines.

ML (Machine Learning): A smaller doll inside AI, representing one method for achieving artificial intelligence by teaching machines to learn from data.

DL (Deep Learning): The innermost doll, representing an advanced, specific type of machine learning that uses complex, multi-layered neural networks.

Data Science is not a doll inside this sequence. Instead, it is the toolkit or workshop where these dolls (ML and DL algorithms) are used to solve real-world problems and find meaningful insights from data.

Artificial Intelligence (AI)

AI is the broad field of computer science dedicated to creating systems that can perform tasks that typically require human intelligence, such as learning, reasoning, problem-solving, and perception.

Goal: To simulate human intelligence in machines to perform intelligent tasks.

Scope: The broadest of the four terms, encompassing everything from simple rule-based systems to complex neural networks.

Techniques: Can include machine learning, deep learning, natural language processing, computer vision, and expert systems.

Examples: Virtual assistants like Siri and Alexa, autonomous vehicles, and chess-playing computers.

Machine Learning (ML)

ML, a subset of AI, focuses on teaching computers to learn from data. It aims to identify patterns for predictions and is narrower in scope than AI, concentrating on data learning. Techniques include supervised, unsupervised, and reinforcement learning. Examples include recommendation systems and spam filters.

Deep Learning (DL)

DL is a subset of ML using multi-layered neural networks to process data and find complex patterns. Its goal is to solve more complex problems, especially with large datasets. DL is more specialized than ML, utilizing "deep" neural networks and often requires more data and computational power. Examples include image recognition and self-driving cars.

Data Science (DS)

DS is a multidisciplinary field using scientific methods to extract knowledge from data. Its objective is to derive insights from data to solve real-world issues. DS has a broad scope, combining concepts from various fields. Techniques span the data lifecycle from collection to visualization and applying ML/DL models. Examples include analyzing customer behavior and predicting market trends.

Q.2 - What are the types of machine learning? Describe each with one
real-world example.
  - Machine learning (ML) is broadly categorized into four types, each defined by how a model learns from data. These types are:

Supervised Learning
Unsupervised Learning
Semi-Supervised Learning
Reinforcement Learning

1. Supervised Learning

In this approach, the model is trained on a "labeled dataset," which means that every input has a corresponding correct output. The model learns the mapping from input to output and can then make predictions on new, unseen data.

Real-world example: Email spam detection

Data: A dataset of emails, where each email is labeled as either "spam" or "not spam."

Process: The algorithm is trained to recognize patterns and keywords that indicate spam, such as certain phrases, suspicious links, or specific senders.

Outcome: The model learns to correctly classify new incoming emails as either spam or legitimate.

2. Unsupervised Learning

Unlike supervised learning, this method uses an "unlabeled dataset" where the model is not given correct answers or categories. Its goal is to find hidden patterns, similarities, and structures within the data on its own.

Real-world example: Customer segmentation

Data: A dataset of customer purchasing habits, demographics, and browsing history, but without pre-defined labels for customer types.

Process: An unsupervised learning algorithm, such as a clustering algorithm, groups the customers into segments based on shared characteristics (e.g., "bargain hunters" versus "high-spending, frequent shoppers").

Outcome: The company gains insight into its customer base, allowing it to tailor marketing strategies for each segment.

3. Semi-Supervised Learning

This is a hybrid approach that uses both a small amount of labeled data and a large amount of unlabeled data for training. It is useful in situations where obtaining labeled data is expensive or time-consuming.

Real-world example: Language translation

Data: A small amount of parallel text with human-provided translations (labeled data) and a vast amount of text in each language without translations (unlabeled data).

Process: The model is trained on the labeled data to learn the basics of translation. It then uses the unlabeled text to further improve its understanding of grammar, sentence structure, and vocabulary in both languages.

Outcome: The model leverages the abundant unlabeled data to significantly improve the accuracy of its translations.

Reinforcement Learning

In this method, an agent learns to make decisions by interacting with an environment. Through a system of rewards and penalties, the agent learns the best sequence of actions to maximize its cumulative reward over time.

Real-world example: Training a self-driving car

Agent and environment: The agent is the car's control system, and the environment is the road, including other vehicles, traffic lights, and pedestrians.

Actions, rewards, and penalties: The agent's actions include accelerating, braking, and steering. It is given a positive reward for safely navigating the road and reaching its destination and a penalty for undesirable behaviors like hitting an obstacle or swerving.

Outcome: Through millions of simulated and real-world trials, the car's system learns the optimal policy for making split-second decisions to drive safely.

Q.3 - Define overfitting, underfitting, and the bias-variance tradeoff in machine learning.
  - Overfitting, underfitting, and the bias-variance tradeoff are critical concepts in machine learning that describe a model's performance and ability to generalize to new data.

Overfitting

Overfitting occurs when a model learns the training data too well, including its random noise and outliers, rather than the underlying pattern. An overfit model will have excellent performance on its training data but will perform poorly on new, unseen data because it has memorized the training set instead of generalizing from it.

Behavior: The model is overly complex and flexible, fitting the training data with excessive detail.

Performance: Low error on the training data, but high error on the test data.

Example: A polynomial regression model with a very high degree that creates a highly complex, "wiggly" curve to hit every training data point. This model will not generalize well to new data points that don't follow the exact same fluctuations.

Underfitting

Underfitting happens when a model is too simple to capture the underlying patterns in the training data. This results in poor performance not only on the training data but also on new, unseen data.

Behavior: The model is too simplistic and makes too many assumptions, failing to learn the relationships between input features and output labels effectively.

Performance: High error on both the training and test data.

Example: Using a simple linear regression model to predict a dataset that has a complex, non-linear, or curved relationship between variables. The straight line is too basic to capture the trend.

The Bias-Variance Tradeoff

The bias-variance tradeoff is the central conflict in machine learning that guides model selection and tuning. It is the balancing act between a model's simplicity (high bias, low variance) and its complexity (low bias, high variance) to minimize total prediction error.

1. Bias: The error caused by a model's simplifying assumptions.

High Bias: The model is too simple and makes strong assumptions, leading to underfitting. The predictions are consistent but inaccurate.

Low Bias: The model is more complex and makes fewer assumptions, fitting the training data more closely.

2. Variance: The error caused by a model's sensitivity to small fluctuations in the training data.

High Variance: The model is overly complex and fits the training data, including its noise, leading to overfitting. The predictions are inconsistent and vary significantly based on the training data.

Low Variance: The model is less sensitive to the training data and is therefore more stable.

The Tradeoff:
The total error of a model can be broken down as:
\(\text{Total\ Error}=(\text{Bias})^{2}+\text{Variance}+\text{Irreducible\ Error}\)

The goal is to find the "sweet spot" of model complexity where the total error is minimized.

Increasing model complexity typically decreases bias but increases variance.Decreasing model complexity typically increases bias but decreases variance.

Example:

High-Bias, Low-Variance: A decision tree with very few splits is simple, making consistent but potentially inaccurate predictions (underfitting).

Low-Bias, High-Variance: A decision tree with an unlimited number of splits can become very complex, fitting every data point (including noise). It will be very accurate on the training data but perform poorly on new data (overfitting)

Q.4 - What are outliers in a dataset, and list three common techniques for
handling them.

  - An outlier is a data point that is significantly different from other observations in a dataset. It falls outside the typical range of values and does not follow the general pattern of the data. Outliers can arise from various sources, including measurement errors, data entry mistakes, or they may simply represent a natural, but rare, variation in the population.

Outliers are important to handle correctly because they can skew statistical results and impact the performance of machine learning models. For instance, an outlier can disproportionately affect the mean and standard deviation, leading to misleading conclusions about the data.

Common techniques for handling outliers

Removal or Trimming: The simplest approach is to delete the rows or entries that contain outlier values.

When to use: This method is best suited for when outliers are clearly identified as data entry errors, measurement mistakes, or other anomalies that do not represent valid data. It is also more appropriate for large datasets where removing a few data points will not significantly impact the overall sample size.

Caution: Removing outliers can sometimes lead to a loss of valuable information or introduce bias if they represent genuine, albeit rare, events.

Example: If a dataset of human heights mistakenly includes a value of 10 feet due to a typo, it should be removed.

Winsorization (Capping): This technique replaces the outlier values with the nearest non-outlier value, effectively "capping" the extreme values at a predetermined threshold.

When to use: Use this method when you want to minimize the influence of outliers without removing them entirely. It is a good choice when you suspect the outliers contain some useful information but are too extreme to be used as-is.

How it works: A common approach is to set the caps based on the interquartile range (IQR) method, replacing all values above the upper fence with the value of the upper fence, and all values below the lower fence with the value of the lower fence.

Example: In a salary dataset, a very small number of extremely high earners could skew the average salary. Winsorization would cap these top earners at a certain percentile (e.g., the 95th percentile), reducing their impact on the mean while preserving their general position in the dataset.

Using Robust Statistical Methods: Some statistical measures and algorithms are inherently less sensitive to outliers. Instead of modifying the data, this approach focuses on using analysis methods that are more stable in the presence of extreme values.

When to use: This technique is ideal when you believe the outliers are not errors but genuine, rare occurrences. It allows you to retain all the original data without the risk of biasing your analysis through removal or manipulation.

Examples:

Using the median instead of the mean to measure central tendency, as the median is not affected by extreme values.

Using quantile regression instead of standard linear regression, as it focuses on estimating the conditional median or other quantiles rather than the conditional mean.

Using non-parametric statistical tests, which do not make assumptions about the data's distribution and are thus more robust to outliers.

Q.5 - : Explain the process of handling missing values and mention one
imputation technique for numerical and one for categorical data.

  - The process of handling missing values is a crucial step in data preprocessing, as many machine learning algorithms cannot handle missing data. The general process involves several steps to detect, analyze, and address missing values in a dataset.

  The process of handling missing values

Identify Missing Values: The first step is to identify where and how many missing values are present in the dataset. Missing values can be represented in various ways, such as "NaN," "Null," "None," or other placeholders. Data can be visualized using tools to show the distribution of missingness.

Determine the Cause: Understanding the reason for missing data can guide the imputation strategy. Missingness can be classified into three types:

Missing Completely at Random (MCAR): Missing data is not related to any other variable in the dataset.

Missing at Random (MAR): Missingness is related to other observed variables. For example, older individuals might be less likely to answer a survey question, but their age is known.

Missing Not at Random (MNAR): Missingness is related to the missing value itself. For instance, people with high salaries might be less likely to report their income.

Choose a Strategy: Based on the analysis, a strategy is chosen. The two main approaches are:

Deletion: Removing rows or columns with missing values. This is suitable for very small amounts of missing data but can lead to a loss of information and introduce bias.

Imputation: Replacing missing values with a substituted value. This preserves the sample size and is often a better option.

Imputation techniques

Numerical data: k-Nearest Neighbors (k-NN) Imputation

k-NN imputation is a more sophisticated method for numerical data that replaces missing values based on the values of the "k-nearest" or most similar data points.

How it works:

For a data point with a missing value, the algorithm finds the k other data points in the dataset that are most similar across all other features.

The missing value is then imputed using the average (mean or median) of the values from its k nearest neighbors.

This method leverages the correlations and relationships between features to produce a more accurate and contextual estimation than simpler methods like mean or median imputation.

Categorical data: Mode Imputation

Mode imputation is a simple and common technique for categorical data, where missing values are replaced with the most frequently occurring category in that column.

How it works:

The algorithm scans all the non-missing values in a categorical column to identify the category that appears most often.

It then replaces all missing values with this most frequent category.

Alternative: For cases where the missingness might be informative, a missing value can be replaced with a new category, such as "Unknown," to signal that the data was originally absent.

```python
"""Q.8 - Write a Python script to:
● Generate 1000 samples from a normal distribution.
● Introduce 50 random missing values.
● Fill missing values with the column mean.
● Plot a histogram before and after imputation."""

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Generate 1000 samples from a normal distribution
np.random.seed(42)
data = np.random.normal(loc=0, scale=1, size=1000)

# Step 2: Introduce 50 random missing values
missing_indices = np.random.choice(1000, 50, replace=False)
data_with_missing = data.copy()
data_with_missing[missing_indices] = np.nan

# Step 3: Fill missing values with the column mean
series_with_missing = pd.Series(data_with_missing)
imputed_series = series_with_missing.fillna(series_with_missing.mean())

# Step 4: Plot histogram before and after imputation
plt.figure(figsize=(12, 5))

# Before imputation
plt.subplot(1, 2, 1)
plt.hist(series_with_missing.dropna(), bins=30, edgecolor='black')
plt.title("Before Imputation (Missing Values Removed)")
plt.xlabel("Value")
plt.ylabel("Frequency")

# After imputation
plt.subplot(1, 2, 2)
plt.hist(imputed_series, bins=30, edgecolor='black')
plt.title("After Imputation (Missing Filled with Mean)")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

Q. 10 - You are working as a data scientist for a retail company. You receive a customer
transaction dataset that contains:
● Missing ages,
● Outliers in transaction amount,
● A highly imbalanced target (fraud vs. non-fraud),
● Categorical variables like payment method.
Explain the step-by-step data preparation plan you’d follow before training a machine learning
model. Include how you’d address missing data, outliers, imbalance, and encoding.
(Include your Python code and output in the code box below.)

Data preparation plan

1. Data Splitting (Before Preprocessing)

To prevent data leakage, the data must be split into training and testing sets before any preprocessing steps are applied. This ensures that all imputation, scaling, and encoding operations are learned from the training data only and then applied to both the training and test sets.

2. Handle Missing Data (Missing Ages)

Missing age data can be handled through imputation. Since age is a numerical variable, the median is a robust choice for imputation as it is less sensitive to potential outliers than the mean.

Technique: Median Imputation

Implementation: Replace missing values in the age column with the median age calculated from the training data.

3. Handle Outliers (Transaction Amount)

Outliers in numerical data like transaction amounts can distort a model's training. It's often best to handle them carefully rather than deleting rows, which could lead to data loss.

Technique: Winsorization (Capping)

Implementation: Cap the transaction_amount at a certain percentile (e.g., the 95th percentile). This approach keeps all data points but limits the influence of extreme values.

4. Handle Class Imbalance (Fraud vs. Non-Fraud)

A highly imbalanced target variable (like fraud detection) requires special handling to prevent the model from becoming biased toward the majority class (non-fraud).

Technique: Synthetic Minority Over-sampling Technique (SMOTE)

Implementation: SMOTE creates synthetic samples of the minority class (fraud) to balance the class distribution. The new samples are generated by taking existing minority samples and randomly tweaking them based on their k-nearest neighbors.

5. Encode Categorical Variables (Payment Method)

Categorical variables, such as payment methods, must be converted into a numerical format that machine learning models can process.

Technique: One-Hot Encoding

Implementation: This technique creates a new binary column for each unique payment method. A '1' in a column indicates that the transaction used that specific payment method, while '0' indicates it did not. This avoids giving the model a false sense of order or hierarchy between different payment methods.

6. Feature Scaling (Transaction Amount and Age)

Scaling numerical features is important for many machine learning algorithms (e.g., those based on gradient descent or distance calculation) to ensure they are all on a comparable scale.

Technique: Standard Scaling

Implementation: Standard Scaling standardizes features by removing the mean and scaling to unit variance. It transforms data such that its distribution has a mean of 0 and a standard deviation of 1.

```python
# =============================================
# Retail Fraud Detection - Data Preparation Plan
# =============================================

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.utils import resample

# -----------------------------
# 1. Simulate a customer dataset
# -----------------------------
np.random.seed(42)
n = 2000

df = pd.DataFrame({
    "age": np.random.randint(18, 70, size=n).astype(float),
    "transaction_amount": np.random.exponential(scale=100, size=n),
    "payment_method": np.random.choice(["CreditCard", "DebitCard", "UPI", "Wallet"], size=n),
    "is_fraud": np.random.choice([0, 1], size=n, p=[0.98, 0.02])  # 2% fraud
})

# Inject ~15% missing ages
missing_idx = np.random.choice(df.index, size=int(0.15 * n), replace=False)
df.loc[missing_idx, "age"] = np.nan

# Inject ~1% extreme outliers in transaction amount
outlier_idx = np.random.choice(df.index, size=int(0.01 * n), replace=False)
df.loc[outlier_idx, "transaction_amount"] *= 50

print("=== Initial Dataset Info ===")
print("Shape:", df.shape)
print("Class distribution:\n", df["is_fraud"].value_counts())
print("Missing ages:", df["age"].isna().sum())
print("Extreme outliers injected:", len(outlier_idx))
print("\nSample rows:\n", df.head(), "\n")

# -----------------------------
# 2. Train/Test Split
# -----------------------------
X = df.drop(columns="is_fraud")
y = df["is_fraud"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

print("Train class distribution:\n", y_train.value_counts())
print("Test class distribution:\n", y_test.value_counts(), "\n")

# -----------------------------
# 3. Feature Engineering
# -----------------------------
# Age: add missing indicator
X_train["age_missing"] = X_train["age"].isna().astype(int)
X_test["age_missing"] = X_test["age"].isna().astype(int)

# Transaction amount: log-transform
X_train["transaction_amount"] = np.log1p(X_train["transaction_amount"])
X_test["transaction_amount"] = np.log1p(X_test["transaction_amount"])

# Amount extreme flag (based on training 99th percentile)
thresh = X_train["transaction_amount"].quantile(0.99)
X_train["amount_extreme"] = (X_train["transaction_amount"] > thresh).astype(int)
X_test["amount_extreme"] = (X_test["transaction_amount"] > thresh).astype(int)

# -----------------------------
# 4. Preprocessing Pipeline
# -----------------------------
numeric_features = ["age", "transaction_amount"]
categorical_features = ["payment_method"]
other_features = ["age_missing", "amount_extreme"]

numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])
categorical_transformer = OneHotEncoder(handle_unknown="ignore")

preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numeric_features),
        ("cat", categorical_transformer, categorical_features)
    ],
    remainder="passthrough"  # keep other engineered features
)

# Fit transformer on train, apply to both train/test
X_train_prep = preprocessor.fit_transform(X_train)
X_test_prep = preprocessor.transform(X_test)

# -----------------------------
# 5A. Model with Class Weights
# -----------------------------
clf_weighted = RandomForestClassifier(class_weight="balanced", random_state=42)
clf_weighted.fit(X_train_prep, y_train)
pred_w = clf_weighted.predict(X_test_prep)

print("=== RandomForest (class_weight='balanced') ===")
print(classification_report(y_test, pred_w, digits=4))
print("Confusion Matrix:\n", confusion_matrix(y_test, pred_w), "\n")

# -----------------------------
# 5B. Oversampling Minority
# -----------------------------
X_train_bal, y_train_bal = resample(
    X_train[y_train == 1], y_train[y_train == 1],
    replace=True, n_samples=y_train.value_counts()[0], random_state=42
)
X_train_over = pd.concat([X_train, X_train_bal])
y_train_over = pd.concat([y_train, y_train_bal])

# Preprocess oversampled set
X_train_over_prep = preprocessor.fit_transform(X_train_over)
clf_over = RandomForestClassifier(random_state=42)
clf_over.fit(X_train_over_prep, y_train_over)

pred_o = clf_over.predict(X_test_prep)

print("=== RandomForest (Oversampled Training Data) ===")
print(classification_report(y_test, pred_o, digits=4))
print("Confusion Matrix:\n", confusion_matrix(y_test, pred_o))
```
